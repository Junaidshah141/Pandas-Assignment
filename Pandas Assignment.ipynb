```python
import pandas as pd
```

```python
dt = pd.read_csv('Pokemon.csv')
dt
```

```python
dt["Type 1"].value_counts()
```

```python
dt.sort_values("Speed", ascending=False)
```

```python
attack_mean = dt["Attack"].mean()
def set_attack(val):
    if val < attack_mean:
        return "Attack_low"
    else:
        return "Attack_high"
    
dt["Attack_High_Low"] = dt["Attack"].apply(set_attack)
dt
```

```python
speed_mean = dt["Speed"].mean()
def set_speed(value):
    if value < speed_mean:
        return "Speed id low"
    else:
        return "Speed is high"
dt["High_Low_Speed"] = dt["Speed"].apply(set_speed)
dt
```

```python
HP_mean = dt["HP"].mean()
def set_HP(x):
    if x < HP_mean:
        return "HP is low"
    else:
        return "HP is high"
dt["High_Low_HP"] = dt["HP"].apply(set_HP)
dt
```
